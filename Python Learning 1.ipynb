```python
# Input, higher or lower with count guesses / games

import random
import time
print("Please guess a number between 1 and 100")
time.sleep(3)
print("Picking a number...")
time.sleep(2)

guess = int(input("What is your guess?: "))
correct_number = random.randint(1,100)
guess_count = 0

while guess != correct_number:
  time.sleep(1)
  guess_count += 1
  if guess < correct_number:
    guess = int(input("Please try again with bigger number!: "))
  else:
     guess = int(input("Please try again with smaller number!: "))1

print(f"You got the right answer, it's {correct_number}. It took you {guess_count} guesses. ")
```
